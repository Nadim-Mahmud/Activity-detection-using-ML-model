```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

import glob
import os
```

# Reading data

```python
FILE_PATH = "DataSet2/**/*.csv"
```

```python
# Read fileList from dataSet1 using glob
file_path_list = glob.glob(FILE_PATH)
```

```python
sensor_data = pd.DataFrame(file_path_list, columns = ['path'])

sensor_data.head()
```

```python
# Parse file name

sensor_data["file_name"] = sensor_data["path"].apply(lambda x: os.path.basename(x))

sensor_data.head()
```

```python
#Parse experiment id and store into int form

def parse_exp_id(file_name):
    exp_id = file_name.split('_')[0]
    
    return int(''.join(filter(str.isdigit, exp_id)))


sensor_data['exp_id'] = sensor_data['file_name'].apply(lambda x: parse_exp_id(x))

sensor_data.head()
```

```python
sensor_data['exp_id'].value_counts()
```

```python
#Parse sensore data

sensor_data['sensor'] = sensor_data['file_name'].apply(lambda x: x.split("_")[4])
sensor_data['frequency'] = sensor_data['file_name'].apply(lambda x: x.split("_")[5])

sensor_data.head()
```

```python
# Activity dictionary

activity_dict = {
    1:  'Sitting',
    2:  'Sitting',
    3:  'Using computer',
    4:  'Using computer',
    5:  'While sitting',
    6:  'While sitting',
    7:  'Sitting',
    8:  'Standing',
    9:  'Standing',
    10: 'Standing',
    11: 'Standing',
    12: 'Sitting',
    13: 'Standing',
    14: 'Sitting',
    15: 'Standing',
}

# Activity details dictionary

activity_details_dict = {
    1:  'Reading a book',
    2:  'Writing in a notebook',
    3:  'Typing',
    4:  'Browsing',
    5:  'Moving head, body',
    6:  'Moving chair',
    7:  'Stand up from sitting',
    8:  'Standing',
    9:  'Walking',
    10: 'Running',
    11: 'Taking stairs',
    12: 'Stationary->Wear it->Put it back->stationary (15 s delay between thetransitions)',
    13: 'Stationary->Wear it->Put it back->stationary(15 s delay between the transitions)',
    14: 'Pick up items from floor',
    15: 'Pick up items from floor'
}
```

```python
# calculte experiment no
# total activity 15. rounding each number and +1 to adjust 1-15 series
# each missing value and exp id inconsistancy resolved by each if conditions 

def calc_activity_no(exp_id):
    
    if exp_id < 307:
        
        return ((exp_id + 2) % 15 + 1)
    elif exp_id < 320:
        
        return ((exp_id + 3) % 15 + 1)
    elif exp_id < 334:
        
        return ((exp_id + 4) % 15 + 1)
    elif exp_id < 348:
        
        return ((exp_id + 5) % 15 + 1)
    else:
        return ((exp_id + 6) % 15 + 1)
    
def get_activity(activity_id):
    return activity_dict[activity_id]

def get_activity_details(activity_id):
    return activity_details_dict[activity_id]
```

```python
sensor_data['activity_id'] = sensor_data['exp_id'].apply(lambda x: calc_activity_no(x))
sensor_data['activity'] = sensor_data['activity_id'].apply( lambda x: get_activity(x))
sensor_data['activity_details'] = sensor_data['activity_id'].apply( lambda x: get_activity_details(x))

print(sensor_data.shape)
sensor_data.head()
```

# D A T A  &nbsp; C L E A N I N G

Discarding whole exeperiemnt if either of accelerometer or gyroscope data is not present

```python
feature_of_interest = ['Accelerometer', 'Gyroscope']

sensor_data = sensor_data.groupby('exp_id').filter(
    lambda x: all(elmnt in x['sensor'].values for elmnt in feature_of_interest)
)
```

```python
print(sensor_data.shape)
sensor_data.to_csv('sensor_data.csv')
```

# V I S U A L I Z A T I O N

Visualization helper functions

```python
def read_acc_gyro_mag_data(path):
    df = pd.read_csv(path)
    df.columns = ['epoch', 'time', 'elapsed', 'x', 'y', 'z']
    
    return df
```

```python
def show_sensore_fig(data, label):
    fig, ax = plt.subplots(1, 1, figsize=(8, 6), sharex = True)
    
    ax.plot(data['elapsed'], data['x'], label='x')
    ax.plot(data['elapsed'], data['y'], label='y')
    ax.plot(data['elapsed'], data['z'], label='z')
    
    plt.title(label)
    plt.show()
```

```python
def show_pressure_figure(path, label):
    data = pd.read_csv(path)
    
    plt.figure(figsize=(8, 6))
    plt.plot(data['elapsed (s)'], data['pressure (Pa)'], label='pressure (Pa)')
    plt.title(label)
    plt.show()
```

# WINDOWING & FEATURE EXTRACTION

```python
sensor_data.head()
```

Feature extraction for accellerometer data

```python
WINDOW_SIZE = 100
STEP_SIZE = 50
```

```python
feature_list = []

for exp_id in sensor_data['exp_id'].value_counts().index:
    temp_df = sensor_data[sensor_data['exp_id'] == exp_id]
    
    acc_df = pd.read_csv(temp_df[temp_df['sensor'] == 'Accelerometer'].iloc[0 , 0])
    acc_df.columns = ['epoch', 'time', 'elapsed', 'x', 'y', 'z']
    acc_df = acc_df[['x', 'y', 'z']]

    gyro_df = pd.read_csv(temp_df[temp_df['sensor'] == 'Gyroscope'].iloc[0 , 0])
    gyro_df.columns = ['epoch', 'time', 'elapsed', 'x', 'y', 'z']
    gyro_df = acc_df[['x', 'y', 'z']]

      
    for i in range(0, len(acc_df) - WINDOW_SIZE, STEP_SIZE):
        window_acc_data = acc_df.iloc[i : i + WINDOW_SIZE]

        #computing acc features
        acc_x_mean = window_acc_data['x'].mean()
        acc_y_mean = window_acc_data['y'].mean()
        acc_z_mean = window_acc_data['z'].mean()
        
        acc_x_median = window_acc_data['x'].median()
        acc_y_median = window_acc_data['y'].median()
        acc_z_median = window_acc_data['z'].median()

        acc_x_var = window_acc_data['x'].var()
        acc_y_var = window_acc_data['y'].var()
        acc_z_var = window_acc_data['z'].var()

        acc_x_std = window_acc_data['x'].std()
        acc_y_std = window_acc_data['y'].std()
        acc_z_std = window_acc_data['z'].std()
        
        acc_x_min = window_acc_data['x'].min()
        acc_y_min = window_acc_data['y'].min()
        acc_z_min = window_acc_data['z'].min()
        
        acc_x_max = window_acc_data['x'].max()
        acc_y_max = window_acc_data['y'].max()
        acc_z_max = window_acc_data['z'].max()

        acc_x_max_min_diff = window_acc_data['x'].max() - window_acc_data['x'].min()
        acc_y_max_min_diff = window_acc_data['y'].max() - window_acc_data['y'].min()
        acc_z_max_min_diff = window_acc_data['z'].max() - window_acc_data['z'].min()

        acc_x_iqr = window_acc_data['x'].quantile(0.75) - window_acc_data['x'].quantile(0.25)
        acc_y_iqr = window_acc_data['y'].quantile(0.75) - window_acc_data['y'].quantile(0.25)
        acc_z_iqr = window_acc_data['z'].quantile(0.75) - window_acc_data['z'].quantile(0.25)

        acc_x_skew = window_acc_data['x'].skew()
        acc_y_skew = window_acc_data['y'].skew()
        acc_z_skew = window_acc_data['z'].skew()

        #computing gyro features
        window_gyro_data = gyro_df.iloc[i : i + WINDOW_SIZE]

        gyro_x_mean = window_gyro_data['x'].mean()
        gyro_y_mean = window_gyro_data['y'].mean()
        gyro_z_mean = window_gyro_data['z'].mean()
        
        gyro_x_median = window_gyro_data['x'].median()
        gyro_y_median = window_gyro_data['y'].median()
        gyro_z_median = window_gyro_data['z'].median()

        gyro_x_var = window_gyro_data['x'].var()
        gyro_y_var = window_gyro_data['y'].var()
        gyro_z_var = window_gyro_data['z'].var()

        gyro_x_std = window_gyro_data['x'].std()
        gyro_y_std = window_gyro_data['y'].std()
        gyro_z_std = window_gyro_data['z'].std()
        
        gyro_x_min = window_gyro_data['x'].min()
        gyro_y_min = window_gyro_data['y'].min()
        gyro_z_min = window_gyro_data['z'].min()
        
        gyro_x_max = window_gyro_data['x'].max()
        gyro_y_max = window_gyro_data['y'].max()
        gyro_z_max = window_gyro_data['z'].max()
        
        gyro_x_max_min_diff = window_gyro_data['x'].max() - window_gyro_data['x'].min()
        gyro_y_max_min_diff = window_gyro_data['y'].max() - window_gyro_data['y'].min()
        gyro_z_max_min_diff = window_gyro_data['z'].max() - window_gyro_data['z'].min()

        gyro_x_iqr = window_gyro_data['x'].quantile(0.75) - window_gyro_data['x'].quantile(0.25)
        gyro_y_iqr = window_gyro_data['y'].quantile(0.75) - window_gyro_data['y'].quantile(0.25)
        gyro_z_iqr = window_gyro_data['z'].quantile(0.75) - window_gyro_data['z'].quantile(0.25)

        gyro_x_skew = window_gyro_data['x'].skew()
        gyro_y_skew = window_gyro_data['y'].skew()
        gyro_z_skew = window_gyro_data['z'].skew()

        feature_list.append([acc_x_mean, 
                             acc_y_mean, 
                             acc_z_mean,
                             acc_x_median,
                             acc_y_median,
                             acc_z_median,
                             acc_x_var, 
                             acc_y_var, 
                             acc_z_var, 
                             acc_x_std, 
                             acc_y_std, 
                             acc_z_std,
                             acc_x_min,
                             acc_y_min,
                             acc_z_min,
                             acc_x_max,
                             acc_y_max,
                             acc_z_max,
                             acc_x_max_min_diff,
                             acc_y_max_min_diff,
                             acc_z_max_min_diff,
                             acc_x_iqr,
                             acc_y_iqr,
                             acc_z_iqr,
                             acc_x_skew,
                             acc_y_skew,
                             acc_z_skew,
                             gyro_x_mean, 
                             gyro_y_mean, 
                             gyro_z_mean,
                             gyro_x_median,
                             gyro_y_median,
                             gyro_z_median,
                             gyro_x_var, 
                             gyro_y_var, 
                             gyro_z_var, 
                             gyro_x_std, 
                             gyro_y_std, 
                             gyro_z_std,
                             gyro_x_min,
                             gyro_y_min,
                             gyro_z_min,
                             gyro_x_max,
                             gyro_y_max,
                             gyro_z_max,
                             gyro_x_max_min_diff,
                             gyro_y_max_min_diff,
                             gyro_z_max_min_diff,
                             gyro_x_iqr,
                             gyro_y_iqr,
                             gyro_z_iqr,
                             gyro_x_skew,
                             gyro_y_skew,
                             gyro_z_skew,
                             temp_df[temp_df['sensor'] == 'Accelerometer'].iloc[0, 5]])
```

```python
feature_df = pd.DataFrame(feature_list)
print(feature_df.head())
```
