```python
import pandas as pd
import numpy as np
import matplotlib as plt

import glob
```

```python
BASE = "DataSet2/**/*.csv"
```

```python
accelerometer_df = pd.DataFrame()
gyroscope_df = pd.DataFrame()
magnetometer_df = pd.DataFrame()
pressure_df = pd.DataFrame()
```

```python
#Read Data

def readAndConcatData(fileName, container):
    df = pd.read_csv(fileName)
    
    splittedDirectory = fileName.split('/')
    user = splittedDirectory[1]
    experiment = splittedDirectory[2].split('_')[0]
    experimentNumber = experiment.search(r'\d+', s).group()
        
    df['user'] = user
    df['experiment'] = experimentNumber
    
    return pd.concat([container, df], ignore_index=True)
    


for fileName in glob.glob(BASE, recursive=True):
    print(f"reading : {fileName}")

    
    if 'Accelerometer' in fileName:
        accelerometer_df = readAndConcatData(fileName, accelerometer_df)
        
    elif 'Gyroscope' in fileName:
        gyroscope_df = readAndConcatData(fileName, gyroscope_df)
        
    elif 'Magnetometer' in fileName:
        magnetometer_df = readAndConcatData(fileName, magnetometer_df)
    
    elif 'Pressure' in fileName:
        pressure_df = readAndConcatData(fileName, pressure_df)
        
```

```python
# Write data (Accelerometer & Gyroscope)

print("Writing data ...")

accelerometer_df.to_csv("accelerometer.csv")
gyroscope_df.to_csv("gyroscope.csv")
```
